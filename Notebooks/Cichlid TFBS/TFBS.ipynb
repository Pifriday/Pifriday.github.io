### Introduction

In both fishes and terrestrial vertebrates mechanoreception is mediated by hair cells. Bundles of actin-filled structures known as stereocilia and kinocilia protrude from the apical surface of hair cells. Ciliary bundles consist of a single kinocilium and multiple stereocilia; stereocilia are connect by tip links and arranged by height with the tallest being linked to the kinocilium. Pressure and movement can cause stereocilia to pivot which opens transduction channels at the tip link's base and allows K+ ions from the surrounding endolymph fluid to enter the cell. When stereocilia are deflected toward the kinocilium, the subsequent ion influx depolarizes the cell’s membrane and causes calcium channels at synaptic end of the cell to open. Synaptic vesicles then fuse to the membrane and neurotransmitters are released. Through this process, hair cells detect both sound and movement in the environment.

Fishes have hair cells in their inner ears and lateral lines. The lateral line is a series of sensory patches called neuromasts that runs along the body of fish. Neuromasts sense both pressure waves and particle displacement in surrounding water; they greatly enhance fish’s ability to school, avoid predators, and detect prey. Inner ear hair cells lie within three organs called the saccule, lagena and utricle. These organs also contain otoliths, calcium carbonate structures that are comparatively denser than the rest of the body. As otoliths move, they deflect stereocilia which causes auditory transduction. While the lateral line is typically associated with movement detection and the inner ear with sound, both systems can mediate either process (Higgs & Radford 2013).
    
Tip links are primarily composed of the cadherin proteins protocadherin 15 (PCDH15) and cadherin 23 (CDH23). PCDH15 connects to the mechanotransduction channels and the shorter stereocilium. CDH23 forms a complex with the proteins of the taller stereocilium: harmonin (USH1C), sans (USH1G), and Myosin VIIA (MYO7A). USH1C and USH1G are implicated in usher syndrome an autosomal recessive disease that causes hearing loss and blindness (Weil et al 2003). Both proteins are required for proper ciliary development and hearing. While these genes are phylogenetically unrelated, USH1G does have a paralog in the gene Ankyrin Repeat And Sterile Alpha Motif Domain Containing 4B (ANKS4B). ANKS4B is found in intestinal microvilli and functions similarly to USH1G; it forms a complex with harmonin, MYO7B, and PCDH24. Structural similarities between gut microvilli and stereocilia implies that they share an evolutionary origin. As development of the gut precedes the evolution of the auditory system, it’s possible that stereocilia evolved from microvilli (Pelaseyed & Bretscher 2018).
    
The primary hypothesis addressed in this report is that the promoters of ANKS4B and USH1G will have binding sites for most of the same transcription factors; and, that USH1G promoters will have more binding sites for transcription factors that are important components of hair cell differentiation. The reasoning behind this hypothesis is that because USH1G and ANKS4B are paralogs and essentially have the same function, they will use many of the same transcription factors. However, because the environment of the intestine is very different from that of the inner ear and lateral line, the genes should differ greatly in how much they depend on transcription factors that are specialized for each setting. The specific transcription factors being analyzed here are Atoh1, Neurog1, Sox2, and Sox3. These transcription factors regulate the differentiation of stem cells into hair cells (Gálvez et al 2017).
    
The transcription factor binding site (TFBS) analysis was performed by scanning the promoters of USH1G and ANKS4B in 5 different cichlid genomes: Oreochromis niloticus, Neolamprologus brichardi, Maylandia zebra, Pundamilia nyererei, and Haplochromis burtoni. Zebrafish are known to have two orthologs of USH1G in USH1GA and USH1GB while humans have only a single USH1G copy (Blanco-Sanchez et al 2018). Presumably, cichlids would have both USH1GA/B like zebrafish. Thus, the second hypothesis addressed in this report is that each cichlid species will have copies of ANKS4B, USH1GA, and USH1GB. 

### Methods

Amino acid sequences for each gene and species combination were acquired using BLASTp and saved to FASTA files. Cichlid copies of ANKS4B, USH1GA, and USH1GB were located using zebrafish sequences from NCBI. Sequences for the aforementioned genes along with the zebrafish sequences for ANKS3/6 and the human USH1G were aligned using ClustalW. To confirm that the correct genes were identified, a consensus tree was created using these aligned sequences. This tree was constructed using the PAM90 similarity scoring matrix and a bootstrap value of 100. This tree was rooted at its midpoint and ladderized so that deeper clades were displayed closer to the top of the tree. To perform the TFBS analysis, promoter sequences for each gene and cichlid/species combination were manually extracted from NCBI. Sequences were scanned for every transcription within the JASPAR CORE database. While being scanned, sequences were truncated to 1000 bp upstream and a 90% relative score cutoff was used. After TFBS data was collected, the data was aggregated find the number of binding sites for each promoter sequence.

```python
from Bio import SeqIO
from Bio.Alphabet import generic_protein, generic_dna, IUPAC

accessions = {} # create a dictionary of genes and their accession numbers
for i in SeqIO.parse("usher_proteins.txt", "fasta", generic_protein):
    des = i.description
    taxa = des[des.find('[')+1:des.find(']')]
    taxa = taxa[0] + '.' + taxa[taxa.find(' '):]
    gene = des[des.find('('):des.find(')')+1]
    new_des = taxa + ' ' + gene
    accessions[des[:des.find(' ')]] = new_des

accessions
```

```python
from Bio.Align.Applications import ClustalwCommandline
from Bio import AlignIO

# algin the sequences
clustalw_exe = r"C:\Program Files (x86)\ClustalW2\clustalw2.exe"
cline = ClustalwCommandline(clustalw_exe, infile="usher_proteins.txt", outfile='usher_clustal.aln')
cline();

# import alignment
with open("usher_clustal.aln", "r") as aln:
    alignment = AlignIO.read(aln, "clustal")
```

```python
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor, DistanceCalculator
from Bio.Phylo.Consensus import bootstrap_consensus, majority_consensus

#construct consensus tree
calculator = DistanceCalculator('pam90')
constructor = DistanceTreeConstructor(calculator)

usher_boot = bootstrap_consensus(alignment, 100, constructor, majority_consensus)
```

```python
from Bio import Phylo
import matplotlib.pyplot as plt

# match accession numbers to genes and species
for i in usher_boot.find_clades():
    if i.name:
        i.name = accessions[i.name]
    elif i.confidence:
        i.name = str(round(i.confidence))
        i.confidence = ''

usher_boot.root_at_midpoint() # root the tree
usher_boot.ladderize() # ladderize the tree
```

```python
Phylo.draw(usher_boot, axes=plt.figure(figsize=(25,12.5), dpi=200).add_subplot(1, 1, 1))
```

```python
from Bio.Seq import Seq
import pandas as pd
from Bio import motifs

# scan the promoter sequences for transcription factor binding sites
sites = pd.DataFrame(columns=['Species','Gene','TF','Position','Score','Relative Score'])

try: # loop takes 30 minutes to run. Import from csv if finished already
    sites = pd.read_csv('sites.csv') 
except:
    for i in SeqIO.parse("usher_promoters.txt", "fasta", generic_dna): # promoter
        promoter = Seq(str(i.seq)[-1000:], IUPAC.unambiguous_dna) # limit to 1kb
        des = str(i.description)
        species = accessions[des[:des.find(' ')]]
        for motif in motifs.parse(open("jaspar_core.txt"), "jaspar"):
            motif.pseudocounts = motifs.jaspar.calculate_pseudocounts(motif)
            pssm = motif.pssm
            max_score = pssm.max
            min_score = pssm.min
            abs_score_threshold = (max_score - min_score) * 0.9 + min_score
            for position, score in pssm.search(promoter, threshold=abs_score_threshold):
                rel_score = (score - min_score) / (max_score - min_score)
                row = {'Species': species[:species.find('(')],'Gene': species[species.find('(')+1:species.find(')')], 'TF':motif.name,
                   'Position': position, 'Score': score, 'Relative Score': rel_score}
                sites.loc[len(sites)] = row

sites.to_csv('sites.csv', index=False)
sites.head()
```

```python
# show the total number of binding sites, the number of unique TFs, and which TFs had the most binding sites
TF_sites = sites['TF']
print(f'{len(TF_sites)} total binding sites ')
print(f'{len(TF_sites.unique())} different transcription factors')
TF_sites.value_counts()[0:10]
```

```python
ANK_sites = sites[sites['Gene'] == 'ANKS4B']['TF']
print(f'{len(ANK_sites.unique())} different transcription factors for ANKS4B')
ANK_sites.value_counts()[0:10]
```

```python
IGA_sites = sites[sites['Gene'] == 'USH1GA']['TF']
print(f'{len(IGA_sites.unique())} different transcription factors for USH1GA')
IGA_sites.value_counts()[0:10]
```

```python
IGB_sites = sites[sites['Gene'] == 'USH1GB']['TF']
print(f'{len(IGB_sites.unique())} different transcription factors for USH1GB')
IGB_sites.value_counts()[0:10]
```

```python
# show transcription factor diversity for each species. %unique shows how many of that species' TFs weren't in other species

unique_species = sites['Species'].unique().tolist()
for species in unique_species:
    species_TFs = sites[sites['Species'] == species]['TF'].unique().tolist()
    other_TFs = []
    for i in unique_species:
        if i != species:
            other_TFs += sites[sites['Species'] == i]['TF'].unique().tolist()
    x = 0
    y = len(species_TFs)
    for TF in species_TFs:
        if TF not in other_TFs:
            x += 1
    print(f'{species}: {y} TFs, {round((x/y)*100, 2)}% unique')
```

```python
# aggregates data for how many binding sites were present for each gene
counts = pd.DataFrame(columns=['Species','TF','ANKS4B','USH1GA','USH1GB'])

TFs = sites['TF'].unique().tolist()
species = sites['Species'].unique().tolist()
genes = sites['Gene'].unique().tolist()
for TF in TFs:
    for s in species:
        row = {'Species': s,'TF': TF, 'ANKS4B': '','USH1GA': '','USH1GB': ''}
        for g in genes:
            cond1 = sites['Species'] == s
            cond2 = sites['Gene'] == g
            cond3 = sites['TF'] == TF
            df = sites[cond1 & cond2 & cond3]
            row[g] = int(len(df))
        counts.loc[len(counts)] = row

counts['USH1G Total'] = counts['USH1GA'] + counts['USH1GB']
ank_cond = counts['ANKS4B'] != 0; ush_cond = counts['USH1G Total'] != 0
counts = counts[ank_cond | ush_cond] # remove TFs that had no binding sites
counts.sort_values('ANKS4B', ascending=False, inplace=True)
counts.head(10)
```

```python
import matplotlib.style as style
style.use('seaborn-talk')
import seaborn as sns

# calculate the correlation between the number of binding sites
corr = counts['ANKS4B'].astype('float64').corr(counts['USH1G Total'].astype('float64'))

plt.figure(figsize=(24,12), dpi=250)
sns.scatterplot(x='ANKS4B',y='USH1G Total', hue='Species', data=counts)
plt.title(f'TF site counts for ANKS4B and USH1G (r = {round(corr, 3)})')
plt.show()
```

```python
corr = counts['USH1GA'].astype('float64').corr(counts['USH1GB'].astype('float64'))

plt.figure(figsize=(24,12), dpi=250)
sns.scatterplot(x='USH1GA',y='USH1GB', hue='Species', data=counts)
plt.title(f'TF site counts for USH1GA and USH1GB (r = {round(corr, 3)})')
plt.show()
```

### Results

One copy for ANKS4B, USH1GA, and USH1GB was found in each cichlid genome. The majority of the bootstrap values for the consensus tree are in the 40-50 range. The branch leading to the copies of USH1GB in M. zebra and P. nyererei are a noticeable outlier with a bootstrap value of 92. As was expected, the zebrafish copies of ANKS3/6 are the outgroups of ANKS4B and USH1G. Also, the zebrafish copies of ANKS4B, USH1GA, and USH1GB are the outgroups to the cichlid genes and the human copy of USH1G is an outgroup to the fish genes. The positioning of cichlid species does not consistently represent the actual relationships between them; while the ANKS4B and USH1GA genes correctly have O. niloticus as the cichlid outgroup and N. brichardi as the next deepest lineage, USH1GB incorrectly has H. burtoni as the second deepest lineage.

Across every promoter scanned, a total of 9865 transcription sites were identified for 377 different transcription factors. The transcription factors with the most sites were GATA2 and FOXLI with 236 and 233 sites respectively. These two transcription factors were within the top three most common sites for each gene. The number of unique transcription factors differed for each gene with ANKS4B, USH1GA, and USH1GB having 263, 282, and 257 unique transcription factors respectively. Total transcription factor counts also differed for each cichlid species, O. niloticus had the most at 335 while N. brichardi had the least at 280. The other three species all had approximately 300 transcription factors. Despite having similar numbers of transcription factors these three taxa differed somewhat in what transcription factor sites were found. For P. nyererei, 4.03% of its unique transcription factors had no sites in the other four species. O. niloticus had the most unique transcription factors with 10.45% of its transcription factors not appearing in other species.

Transcription factor site counts for each gene and species combination were collected by aggregating the site location data. After removing those transcription factors that had no sites from the dataset, a correlation coefficient of 0.8 was calculated for ANKS4B and USH1G counts. Binding sites were identified in all chichlid species for Neurog1 and Sox3. Atoh1 and Sox2 had sites in every species except P. nyererei. Each of these transcription factors had more sites for USH1GA prmoters than USH1GB promoters. Only three binding sites were identified for USH1GB, 1 for Neurog1 and two for Sox3. The transcription factor and species combinations O. niloticus/Atoh1 and P. nyererei/Sox3 had more ANKS4B sites than USH1GA. For every other combination, there were either more USH1GA sites than ANKS4B or the counts were equal.

### Discussion and conclusion

Despite relatively low bootstrap values on the consensus tree, most of the genes and species are positioned as expected. While ANKS4B and USH1GA/B do appear to rely on many of the same transcription factors, there are differences between them. Interestingly, USH1GA and USH1GB appear to be as dissimilar to each other as each is to ANKS4B. The correlation coefficient for USH1GA/B binding site counts was 0.73 compared to 0.80 for ANKS4B and USH1G. Also, site counts for Atoh1, Neurog1, Sox2, and Sox3 were higher in ANKS4B than USH1B. These results imply that the two USH1G genes could be expressed in different locations. One gene could be expressed more in the lateral line while another could be expressed in the ear. As humans only have a single USH1G copy, it’s possible that terrestrial vertebrates lost a paralog as only one was necessary for proper hair cell function in the ear.

The binding site counts for Atoh1, Neurog1, Sox2, and Sox3 do not differ much between genes. While the cichlid species generally had more binding sites for USH1G than ANKS4B, the counts are too low to definitively say that there is a difference. Also, because USH1GA and USH1GB appear to differ in what transcription factors they rely on, it does not make sense to combine their binding site totals to a single value. After testing the hypothesis that USH1GA will have more binding sites than ANKS4B for the transcription factors of interest, the answer is inconclusive. There is some evidence that USH1GA relies more heavily on these transcription factors, however more data will be needed to reach a conclusion. 

This analysis has highlighted how much diversity there is in the transcription factor binding site profiles of different species. Each gene’s promoter has thousands of binding sites for hundreds of different transcription factors. Not only do these genes differ in what transcription factors they use but species differ as well. Even between the closely related species M. zebra, P. nyererei, H. burtoni, each had binding sites for transcription factors that the others lacked. It would be interesting see how heavily TFBS evolution has contributed to cichlid speciation. For future studies, it may be prudent to analyze the promoters of a wider variety of genes across the cichlid genome. And, to look for general trends in binding site counts rather than focus on a few select transcription factors from the beginning.
